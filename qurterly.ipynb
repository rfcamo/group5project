```python
# Dependencies and Setup
import pandas as pd
import numpy as np
pd.options.mode.chained_assignment = None  # default='warn'

# File to Load (Remember to Change These)
data = "Ray Data/employment_study.csv"

# Read Purchasing File and store into Pandas data frame
df = pd.read_csv(data)
df.head(5)
```

```python
df['Month'] = pd.DatetimeIndex(df['Date']).month
df.head()
```

```python
# create a list of quarters values
quarters = []
for index, row in df.iterrows():
    month = df.iloc[index]["Month"]
    # months 1-3 is Q1, 4-6 is Q2, 7-9 is Q3, 10-12 is Q4
    if month <= 3:
        quarter = 1
    elif month <= 6:
        quarter = 2
    elif month <= 9:
         quarter = 3
    else:
        quarter = 4
    quarters.append(quarter)
```

```python
# create a list of midyear
midyear = []
for index, row in df.iterrows():
    month = df.iloc[index]["Month"]
    # months 1-3 is Q1, 4-6 is Q2, 7-9 is Q3, 10-12 is Q4
    if month <= 3:
        halfyear = 1
    else:
        halfyear = 2
    midyear.append(halfyear)
```

```python
df["Quarter"] = quarters
df["Semister"] = midyear
df
```
