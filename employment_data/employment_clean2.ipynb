```python
# Dependencies and Setup
import pandas as pd
import numpy as np
pd.options.mode.chained_assignment = None  

# File to Load (Remember to Change These)
data = "Ray Data/labour_by_state.xls"
# Read source data and store into Pandas data frame
xls_data = pd.read_excel(data, sheet_name = "Data2")

xls_data.head()
```

```python
# Transpose dataframe
trans_to = xls_data.T
trans_to
```

```python
# Filter dataframe to ouput required data
data_filter = ['Series Type', 'Seasonally Adjusted']
filter_data = trans_to[trans_to[1].isin(data_filter)]
```

```python
filter_data.head(10)
```

```python
# Revert transpose to original 
original_data = filter_data.T

original_data.head()
```

```python
# Remove special characters from column name
original_data.columns=original_data.columns.str.replace('[:,-,>,;,.,1, ]','')
original_data.head()
```

```python
# Choose columns for final data
employment_data = original_data.iloc[:,[0,30,31,32,33,34,35,36,37]]
employment_data.head()
```

```python
# Rename columns for more informative reference 
employment_data_cols = ['Date','Unemployment_Rate_NSW',
                        'Unemployment_Rate_VIC','Unemployment_Rate_QLD',
                        'Unemployment_Rate_SA','Unemployment_Rate_WA',
                        'Unemployment_Rate_TAS','Unemployment_Rate_NT',
                        'Unemployment_Rate_ACT']
employment_data.columns = employment_data_cols
employment_data.head()
```

```python
# Drop unwanted data
employment_data.drop(employment_data.index[:9], inplace=True)
employment_data
```

```python
# Normalize DateTime format
employment_data['Date'] = pd.to_datetime(employment_data['Date']).dt.normalize()
employment_data
```

```python
# Select required date for study
covid_data = employment_data[employment_data['Date'] >= '2018-01-01']
covid_data
```

```python
# Get month from Date column and output it to new column name Month
covid_data['Month'] = pd.DatetimeIndex(covid_data['Date']).month
covid_data['Year'] = pd.DatetimeIndex(covid_data['Date']).year

# Reset index values
covid_data.reset_index(drop=True, inplace=True)
covid_data.head(5)
```

```python
# create a list of quarters values
quarters = []
for index, row in covid_data.iterrows():
    month = covid_data.iloc[index]["Month"]
    # months 1-3 is Q1, 4-6 is Q2, 7-9 is Q3, 10-12 is Q4
    if month <= 3:
        quarter = 1
    elif month <= 6:
        quarter = 2
    elif month <= 9:
         quarter = 3
    else:
        quarter = 4
    quarters.append(quarter)
```
