```python
act_cases = []
nsw_cases = []
nt_cases = []
qld_cases = []
sa_cases = []
tas_cases = []
vic_cases = []
wa_cases = []
```

```python
for index, row in df.iterrows():  
    
    next_value = df.iloc[index+1]["ACT"]
    current_value = df.iloc[index]["ACT"] 
    diff = next_value - current_value 
    act_cases.append(diff)
    
    next_value = df.iloc[index+1]["NSW"]
    current_value = df.iloc[index]["NSW"] 
    diff = next_value - current_value 
    nsw_cases.append(diff)
    
    next_value = df.iloc[index+1]["NT"]
    current_value = df.iloc[index]["NT"] 
    diff = next_value - current_value 
    nt_cases.append(diff)
    
    next_value = df.iloc[index+1]["QLD"]
    current_value = df.iloc[index]["QLD"] 
    diff = next_value - current_value 
    qld_cases.append(diff)
    
    next_value = df.iloc[index+1]["SA"]
    current_value = df.iloc[index]["SA"] 
    diff = next_value - current_value 
    sa_cases.append(diff)
    
    next_value = df.iloc[index+1]["TAS"]
    current_value = df.iloc[index]["TAS"] 
    diff = next_value - current_value 
    wa_cases.append(diff)
    
    next_value = df.iloc[index+1]["VIC"]
    current_value = df.iloc[index]["VIC"] 
    diff = next_value - current_value 
    wa_cases.append(diff)
    
    next_value = df.iloc[index+1]["WA"]
    current_value = df.iloc[index]["WA"] 
    diff = next_value - current_value 
    wa_cases.append(diff)
#         print(f"current index: {index}, next index: {index+1}, Current value: {current_value}, Next value: {next_value}, diff: {diff}")
    wa_cases.append(diff)
    print("next date")
```
