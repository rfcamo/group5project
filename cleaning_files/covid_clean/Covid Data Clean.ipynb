```python
# import modules
import pandas as pd
import matplotlib.pyplot as plt
from datetime import datetime,date
```

```python
# retrieving data sets
# cumulative daily cases of covid global
cases_data = "../../Resources/covid_raw/time_series_covid19_confirmed_global.csv"
cases_data = pd.DataFrame(pd.read_csv(cases_data))
# data pre covid for australia
precovid_data = "../../Resources/covid_raw/time_series_precovid19_australia.csv"
precovid_data = pd.DataFrame(pd.read_csv(precovid_data))
cases_data.head() 
```

```python
# segmented data to australian only
australian_cases = cases_data.loc[cases_data["Country/Region"] == 'Australia', :]

australian_cases
```

```python
precovid_data
```

```python
# now that we have australian data we can drop irrelevant columns
australian_cases = australian_cases.drop(["Country/Region","Lat", "Long"], axis = 1)
```

```python
# rename to state as provinces aren't in Australia
australian_cases = australian_cases.rename(columns={"Province/State" : "State"})
precovid_data = precovid_data.rename(columns={"Province/State" : "State"})

australian_cases.head()
```

```python
precovid_data.head(10)
```

```python
# melting date columns into a single one
postcovid_melt = australian_cases.melt(id_vars=["State"],
                     var_name ="Date", value_name="Covid_Cases")
precovid_data_melt = precovid_data.melt(id_vars=["State"],
                     var_name ="Date", value_name="Covid_Cases")

postcovid_melt 
```

```python

precovid_data_melt
```

```python
# appending melt data
australian_cases_melt = precovid_data_melt.append(postcovid_melt )
australian_cases_melt= australian_cases_melt.reset_index(drop=True)
australian_cases_melt
```

```python
# converting "Date" values into a datetime datatype
australian_cases_melt["Date"] = pd.to_datetime(australian_cases_melt["Date"], errors='coerce')
```

```python
# renaming Covid Cases column to be clearer that it acculumates over time
australian_cases_melt = australian_cases_melt.rename(columns={"Covid_Cases": "Covid_Cases_Cumulative"})
australian_cases_melt

```

```python
# segmenting data by state

act_cases = australian_cases_melt[australian_cases_melt["State"] == "Australian Capital Territory"]
nsw_cases = australian_cases_melt[australian_cases_melt["State"] == "New South Wales"]
nt_cases = australian_cases_melt[australian_cases_melt["State"] == "Northern Territory"]
qld_cases = australian_cases_melt[australian_cases_melt["State"] == "Queensland"]
sa_cases = australian_cases_melt[australian_cases_melt["State"] == "South Australia"]
tas_cases = australian_cases_melt[australian_cases_melt["State"] == "Tasmania"]
vic_cases = australian_cases_melt[australian_cases_melt["State"] == "Victoria"]
wa_cases = australian_cases_melt[australian_cases_melt["State"] == "Western Australia"]


```

```python
# resetting index for the for loop
act_cases = act_cases.reset_index(drop=True)
nsw_cases = nsw_cases.reset_index(drop=True)
nt_cases = nsw_cases.reset_index(drop=True)
qld_cases = qld_cases.reset_index(drop=True)
sa_cases = sa_cases.reset_index(drop=True)
tas_cases = tas_cases.reset_index(drop=True)
vic_cases = vic_cases.reset_index(drop=True)
wa_cases = wa_cases.reset_index(drop=True)





wa_cases
```

```python
# create a for loop to generate daily count of covid cases
wa_covid_case_lst = []
for index, row in wa_cases.iterrows(): 
    try:
        #defining the current days day and next day and finding the difference to get a daily count 
        next_value = wa_cases.iloc[index+1]["Covid_Cases_Cumulative"]
        current_value = wa_cases.iloc[index]["Covid_Cases_Cumulative"] 
        diff = next_value - current_value
        
        wa_covid_case_lst.append(diff)
    except:
        print("Last row")
```

```python
# adding and filling month column
act_cases["Month"] = act_cases["Date"].dt.month
qld_cases["Month"] = qld_cases["Date"].dt.month
nsw_cases["Month"] = nsw_cases["Date"].dt.month
nt_cases["Month"] = nt_cases["Date"].dt.month
sa_cases["Month"] = sa_cases["Date"].dt.month
tas_cases["Month"] = tas_cases["Date"].dt.month
vic_cases["Month"] = vic_cases["Date"].dt.month
wa_cases["Month"] = wa_cases["Date"].dt.month
```

```python
# adding and filling year column for each state
act_cases["Year"] = act_cases["Date"].dt.year
qld_cases["Year"] = qld_cases["Date"].dt.year
nsw_cases["Year"] = nsw_cases["Date"].dt.year
nt_cases["Year"] = nt_cases["Date"].dt.year
sa_cases["Year"] = sa_cases["Date"].dt.year
tas_cases["Year"] = tas_cases["Date"].dt.year
vic_cases["Year"] = vic_cases["Date"].dt.year
wa_cases["Year"] = wa_cases["Date"].dt.year
# outputing cases to csv for each state 
act_cases.to_csv("../../Output/covid_output/act_cases.csv")
qld_cases.to_csv("../../Output/covid_output/qld_cases.csv")
nsw_cases.to_csv("../../Output/covid_output/nsw_cases.csv")
nt_cases.to_csv("../../Output/covid_output/nt_cases.csv")
sa_cases.to_csv("../../Output/covid_output/sa_cases.csv")
tas_cases.to_csv("../../Output/covid_output/tas_cases.csv")
vic_cases.to_csv("../../Output/covid_output/vic_cases.csv")
wa_cases.to_csv("../../Output/covid_output/wa_cases.csv")
```

```python
# appending all state data back into one
australian_daily_cases = act_cases.append(qld_cases.append
                                          (nsw_cases.append
                                           (nt_cases.append
                                            (sa_cases.append
                                             (tas_cases.append
                                              (vic_cases.append
                                               (wa_cases)))))))
#saving to csv
australian_daily_cases.to_csv("../../Output/covid_output/australian_daily_cases.csv")
australian_daily_cases
```

```python
# create a list of quarters values
quarters = []
for index, row in australian_daily_cases.iterrows():
    month = australian_daily_cases.iloc[index]["Month"]
    # months 1-3 is Q1, 4-6 is Q2, 7-9 is Q3, 10-12 is Q4
    if month <= 3:
        quarter = 1
    elif month <= 6:
        quarter = 2
    elif month <= 9:
         quarter = 3
    else:
        quarter = 4
    quarters.append(quarter)
```

```python
# adding the quarters column to data frame and filling it with quarter list
australian_daily_cases["Quarter"] = quarters
australian_daily_cases_quarterly = australian_daily_cases

#saving quarterly to csv
australian_daily_cases_quarterly.to_csv("../../Output/covid_output/australian_daily_cases_quarterly.csv")
australian_daily_cases
```
