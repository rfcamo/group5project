```python
import pandas as pd
#retrieving data
```

```python
cases_data = "Resources/time_series_covid19_confirmed_global_metricdate.csv"
cases_data = pd.DataFrame(pd.read_csv(cases_data))
deaths_data = "Resources/time_series_covid19_deaths_global_metricdate.csv"
deaths_data = pd.DataFrame(pd.read_csv(deaths_data))
recovered_data = "Resources/time_series_covid19_confirmed_global_metricdate.csv"
recovered_data = pd.DataFrame(pd.read_csv(recovered_data))
precovid_data = "Resources/time_series_precovid19_australia.csv"
precovid_data = pd.DataFrame(pd.read_csv(precovid_data))
cases_data.head() 
```

```python
australian_cases = cases_data.loc[cases_data["Country/Region"] == 'Australia', :]
australian_deaths = deaths_data.loc[deaths_data["Country/Region"] == 'Australia', :]
australian_recovered = recovered_data.loc[recovered_data["Country/Region"] == 'Australia', :]
australian_cases
```

```python
australian_cases = australian_cases.drop(["Country/Region","Lat", "Long"], axis = 1)
australian_deaths = australian_deaths.drop(["Country/Region","Lat", "Long"], axis = 1)
australian_recovered = australian_recovered.drop(["Country/Region","Lat", "Long"], axis = 1)

```

```python
australian_cases = australian_cases.rename(columns={"Province/State" : "State"})
australian_deaths = australian_deaths.rename(columns={"Province/State" : "State"})
australian_recovered = australian_recovered.rename(columns={"Province/State" : "State"})
precovid_data = precovid_data.rename(columns={"Province/State" : "State"})
australian_cases
```

```python
precovid_data.head(10)
```

```python
australian_cases = australian_cases.append(precovid_data)
australian_cases = australian_cases.fillna(0).drop_duplicates(subset = "State")
australian_deaths = australian_deaths.append(precovid_data)
australian_deaths = australian_deaths.fillna(0).drop_duplicates(subset = "State")
australian_recovered = australian_recovered.append(precovid_data)
australian_recovered = australian_recovered.fillna(0).drop_duplicates(subset = "State")

australian_cases
```

```python
australian_cases_melt = australian_cases.melt(id_vars=["State"],
                     var_name ="Date", value_name="Covid_Cases") 
australian_deaths_melt = australian_deaths.melt(id_vars=["State"],
                     var_name ="Date", value_name="Covid_Cases") 
australian_recovered_melt = australian_recovered.melt(id_vars=["State"],
                     var_name ="Date", value_name="Covid_Cases") 
australian_cases_melt.head(20)    
```

```python
australian_cases_melt["Date"] = pd.to_datetime(australian_cases_melt["Date"], errors='coerce')
australian_deaths_melt["Date"] = pd.to_datetime(australian_deaths_melt["Date"],errors='coerce')
australian_recovered_melt["Date"] = pd.to_datetime(australian_recovered_melt["Date"],errors='coerce')
```

```python
australian_cases_melt.head(10)
```

```python
australian_cases_melt["Month"] = australian_cases_melt["Date"].dt.month
australian_deaths_melt["Month"] = australian_deaths_melt["Date"].dt.month
australian_recovered_melt["Month"] = australian_recovered_melt["Date"].dt.month
australian_cases_melt.head()
```

```python
australian_cases_melt["Year"] = australian_cases_melt["Date"].dt.year
australian_deaths_melt["Year"] = australian_deaths_melt["Date"].dt.year
australian_recovered_melt["Year"] = australian_recovered_melt["Date"].dt.year
australian_cases_melt.head()
```

```python
cases_groupby = australian_cases_melt.groupby(["State","Year", "Month"]).sum()
deaths_groupby = australian_deaths_melt.groupby(["State","Year", "Month"]).sum()
recovered_groupby = australian_recovered_melt.groupby(["State","Year", "Month"]).sum()
```

```python
#save to output
deaths_groupby.to_csv("Resources/covid_deaths.csv", index=False)

deaths_groupby
```
