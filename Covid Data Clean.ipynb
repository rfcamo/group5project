```python
import pandas as pd
#retrieving data
```

```python
cases_data = "Resources/time_series_covid19_confirmed_global.csv"
cases_data = pd.read_csv(cases_data)
deaths_data = "Resources/time_series_covid19_deaths_global.csv"
deaths_data = pd.read_csv(deaths_data)
recovered_data = "Resources/time_series_covid19_confirmed_global.csv"
recovered_data = pd.read_csv(recovered_data)
cases_data.head() 
```

```python
deaths_data.head()
```

```python
recovered_data.head()
```

```python
cases_data = cases_data.drop(["Lat", "Long"], axis = 1)
deaths_data = deaths_data.drop(["Lat", "Long"], axis = 1)
recovered_data = recovered_data.drop(["Lat", "Long"], axis = 1)
cases_data.head()
```

```python
australian_cases = cases_data.loc[cases_data["Country/Region"] == 'Australia', :]
australian_deaths = deaths_data.loc[deaths_data["Country/Region"] == 'Australia', :]
australian_recovered = recovered_data.loc[recovered_data["Country/Region"] == 'Australia', :]
australian_cases
```

```python
australian_cases_melt = australian_cases.melt(id_vars=["Province/State", "Country/Region"],
                     var_name ="Date", value_name="Covid_Cases") 
australian_deaths_melt = australian_deaths.melt(id_vars=["Province/State", "Country/Region"],
                     var_name ="Date", value_name="Covid_Cases") 
australian_recovered_melt = australian_recovered.melt(id_vars=["Province/State", "Country/Region"],
                     var_name ="Date", value_name="Covid_Cases") 
australian_cases_melt.head()    
```

```python
australian_cases_melt["Date"] = pd.to_datetime(australian_cases_melt["Date"])
australian_deaths_melt["Date"] = pd.to_datetime(australian_deaths_melt["Date"])
australian_recovered_melt["Date"] = pd.to_datetime(australian_recovered_melt["Date"])
```

```python
australian_cases_melt.head(10)
```

```python
australian_cases_melt["Month"] = australian_cases_melt["Date"].dt.month
australian_deaths_melt["Month"] = australian_deaths_melt["Date"].dt.month
australian_recovered_melt["Month"] = australian_recovered_melt["Date"].dt.month
australian_cases_melt.head()
```

```python
australian_cases_melt["Year"] = australian_cases_melt["Date"].dt.year
australian_deaths_melt["Year"] = australian_deaths_melt["Date"].dt.year
australian_recovered_melt["Year"] = australian_recovered_melt["Date"].dt.year
australian_cases_melt.head()
```

```python
cases_groupby = australian_cases_melt.groupby(["Province/State","Country/Region","Year", "Month"]).sum()
deaths_groupby = australian_deaths_melt.groupby(["Province/State","Country/Region","Year", "Month"]).sum()
recovered_groupby = australian_recovered_melt.groupby(["Province/State","Country/Region","Year", "Month"]).sum()
cases_groupby
```
